```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating some random data points
n = 1000
x = np.random.uniform(0,1,1000)

# Creating the histogram
plt.figure(figsize=(10,10))

width = 0.5
histmin = np.floor(0)
histmax = np.ceil(1) + width

bins = np.arange(0,1,width)
plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.title('Distribution of 1000 random data points between 0 and 1')

# Saving & showing the figure
plt.savefig('0209_journal.pdf', format='pdf')

plt.show()
```
