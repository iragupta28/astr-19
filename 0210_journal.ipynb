```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# I decided to use a Gaussian (normal) distribution.
x = np.random.normal(loc=0,scale=1,size=1000)

plt.figure(figsize=(10,10))
plt.hist(x, bins=100, color='pink', edgecolor='blue')

plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs. y using a Gaussian number distribution')

plt.savefig('0210_journal_plot.pdf',format='pdf')

plt.show()
```
