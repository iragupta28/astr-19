```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Making an array x that ranges from [0,1] in 100 steps
x = np.linspace(0,1,100)
 
# Printing the array
print(x)

# Defining functions that return sin and cos
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
# Making a subplot with 1 row, 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting sin(x) in the left panel
ax1.plot(x, sin_function(x), color='pink')
ax1.set_title('x vs. sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Plotting cos(x) in the right panel
ax2.plot(x, cos_function(x), color='purple')
ax2.set_title('x vs. cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

# Saving the figure as a pdf
plt.savefig('0208_journal_plot.pdf', format='pdf')

# Printing the graphs
plt.show()
```
